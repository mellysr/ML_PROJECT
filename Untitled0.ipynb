```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Ridge
from sklearn.metrics import accuracy_score, mean_squared_error,r2_score
```

```python
datas=pd.read_csv("/content/Hitters.csv")
datas=datas.dropna()
dms=pd.get_dummies(datas[["League","NewLeague","Division"]])
y=datas["Salary"]
x_=datas.drop(["Salary","League","NewLeague","Division"],axis=1).astype("float64")
x=pd.concat([x_,dms],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=99)
```

```python
ridge_model=Ridge()
```

```python
params={
    "alpha":[0.1,0.5,1,0.01],
    "solver":['svd','auto','lsqr'],
    "max_iter":[100,200,300,400,500,1000],
}
```

```python
ridge_cv=GridSearchCV(ridge_model,params,cv=3,n_jobs=-1,verbose=2)
```

```python
ridge_cv.fit(x_train,y_train)
```

```python
alpha_=ridge_cv.best_params_["alpha"]
solver_=ridge_cv.best_params_["solver"]
max_iter_=ridge_cv.best_params_["max_iter"]
```

```python
Ridge_tuned=Ridge(alpha=alpha_,solver=solver_,max_iter=max_iter_)
Ridge_tuned.fit(x_train,y_train)
```

```python
predict=Ridge_tuned.predict(x_test)
```

```python
MSE=mean_squared_error(y_test,predict)
RMSE=np.sqrt(MSE)
MAE=mean_absolute_error(y_test,predict)
```

```python
r2_score_=r2_score(y_test,predict)
```

```python
r2_score_
```

```python
MSE=mean_squared_error(y_test,predict)
RMSE=np.sqrt(MSE)
MAE=mean_absolute_error(y_test,predict)
print("\n Metrik sonuçları\n")
print("MSE:",MSE)
print("RMSE:",RMSE)
print("MAE:",MAE)

# Bağımsız değişkenlerin önem sıralaması
print("Bağımsız Değişkenlerin Önem Sıralaması\n")
Importance = pd.DataFrame({"Importance": np.abs(Ridge_tuned.coef_) * 100},
                          index=x_train.columns)
Importance.sort_values(by="Importance", ascending=True).plot(kind="barh", color="r")

plt.xlabel("Variable Importance")
plt.ylabel("Features")
plt.title("Feature Importance from Ridge Regression")
plt.gca().legend_ = None
plt.show()

# Division - Salary Grafiği
print("\nDivision - Salary Grafiği\n")
sns.lineplot(data=datas, x="Division", y="Salary")
plt.title("Division vs Salary")
plt.show()

# Histogram grafiği
print("\nHistogram Grafiği\n")
datas.hist(figsize=(12, 12))
plt.show()
```
